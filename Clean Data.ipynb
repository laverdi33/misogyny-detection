# For Song Lyrics:

```python
first_decades = pd.read_csv('Billboard_Top_100_Data.csv')
first_decades = first_decades.drop(columns={'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6'})
first_decades = first_decades[first_decades.Year >= 1960]

last_decade = pd.read_csv('billboardHot100_1999-2019.csv').dropna()
last_decade = last_decade.drop(columns={'Peak.position'}).merge(
    last_decade[['Artists', 'Name','Peak.position']].groupby(['Artists','Name']).min(), 
    on = ['Artists','Name'], how = 'left')
```

```python
cleaned_values = last_decade[last_decade.Year == 0]
for year in range(2013,2020):
    
    year_values = last_decade[last_decade.Year == year].sort_values(
        'Peak.position').drop_duplicates().reset_index(drop=True)
    cleaned_values = cleaned_values.append(year_values.loc[0:99,])
    
new_line = '\n'    
cleaned_values = cleaned_values.drop(columns={'Peak.position'}).rename(columns={
    'Artists':'Artist', 'Name':'Song Name'}).reset_index(drop=True)
cleaned_values.Year = cleaned_values.Year.astype(int)
cleaned_values['Lyrics'] = [s[s.find(new_line)+1:].replace(new_line,' ') for s in cleaned_values.Lyrics]
```

```python
final_data = first_decades.append(cleaned_values).reset_index(drop=True)
final_data.to_pickle("song_data_all.pkl")
```

```python
cleaned_values = last_decade[last_decade.Year == 0]
for year in range(2000,2020):
    
    year_values = last_decade[last_decade.Year == year].sort_values(
        'Peak.position').drop_duplicates().reset_index(drop=True)
    cleaned_values = cleaned_values.append(year_values.loc[0:99,])
    
new_line = '\n'    
cleaned_values = cleaned_values.drop(columns={'Peak.position'}).rename(columns={
    'Artists':'Artist', 'Name':'Song Name'}).reset_index(drop=True)

cleaned_values.Year = cleaned_values.Year.astype(int)
cleaned_values['Lyrics'] = [s[s.find(new_line)+1:].replace(new_line,' ') for s in cleaned_values.Lyrics]
cleaned_values
```

```python
cleaned_values.to_pickle("song_data_2000s.pkl")
```

# For reddit data

```python
reddit_data = pd.read_csv('final_labels.csv')
reddit_data = reddit_data[['body', 'level_1', 'level_2', 'level_3','split']].reset_index(drop=True)
reddit_data
```

```python
class_map = reddit_data[['level_1','level_2','level_3']].drop_duplicates(
                ).sort_values(['level_1','level_2','level_3']).reset_index(drop=True)
#class_map['new_class'] = ['Derogation-Discredit',
#                          'Derogation-Discredit',
#                          'Derogation-Discredit', 
#                          'Objectification-Stereotyping',
#                          'Non',
#                          'Derailing-Attacks',
#                          'Derailing-Attacks',
#                          'Derailing-Attacks',
#                          'Dominance',
#                          'Dominance',
#                          'Dominance',
#                          'Objectification-Stereotyping',
#                          'Threatening-Harassment',
#                          'Threatening-Harassment',
#                          'Threatening-Harassment',
#                          'Non',
#                          'Non',
#                          'Non',
#                          'Non',
#                          'Non']

class_map['new_class'] = ['Derogation',
                          'Derogation',
                          'Derogation', 
                          'Derogation',
                          'Non',
                          'Derogation',
                          'Derogation',
                          'Derogation',
                          'Treatment',
                          'Treatment',
                          'Treatment',
                          'Treatment',
                          'Treatment',
                          'Treatment',
                          'Treatment',
                          'Non',
                          'Non',
                          'Non',
                          'Non',
                          'Non']

reddit_data_class = reddit_data.merge(class_map, on=['level_1','level_2','level_3'], how='left')

class_map
```

```python
# drop duplicates here is for multiple perjoratives in one bit of text
reddit_data_class = reddit_data_class.drop_duplicates().reset_index(drop=True)
```

```python
# still a few duplicates to get rid of...first, some data put in train and test, we will put just in train
reddit_data_clean_1 = reddit_data_class[['body', 'new_class', 'split']].drop_duplicates()
reddit_data_clean_1 = reddit_data_clean_1.merge(reddit_data_clean_1[['body','split']].drop_duplicates(
                        ).groupby(['body']).count().rename(columns={'split':'count'}), on=['body'],how='left')
reddit_data_clean_1 = reddit_data_clean_1[(reddit_data_clean_1['count'] == 1) | 
                                          (reddit_data_clean_1.split == 'train')].reset_index(drop=True)
```

```python
# some more duplicate labels to clean:
reddit_data_clean_2 = reddit_data_clean_1[['body', 'new_class', 'split']]
reddit_data_clean_2 = reddit_data_clean_2.merge(reddit_data_clean_2[['body','new_class']].drop_duplicates(
                        ).groupby(['body']).count().rename(columns={'new_class':'count'}), on=['body'],how='left')
reddit_data_clean_2 = reddit_data_clean_2[(reddit_data_clean_2['count'] == 1) | 
                                          (reddit_data_clean_2.new_class != 'Non')].reset_index(drop=True)
```

```python
# now have final data
final_reddit = reddit_data_clean_2[['body', 'new_class', 'split']].drop_duplicates()
final_reddit['body'] = [" ".join(filter(lambda x:x[0:4]!='http', a.split())) for a in final_reddit['body']]
final_reddit
```

```python
final_reddit.to_pickle("reddit_labels.pkl")
```

# For twitter data

```python
twitter_training = pd.read_csv('en_training.tsv', sep='\t')
twitter_training['split'] = 'train'
twitter_test = pd.read_csv('en_testing.tsv', sep='\t')
twitter_test['split'] = 'test'
twitter_data = twitter_training.append(twitter_test)
twitter_data
```

```python
class_map = twitter_data[['misogynous', 'misogyny_category', 'target']].drop_duplicates(
                ).sort_values(['misogyny_category','target']).reset_index(drop=True)

#class_map['new_class'] = ['Dominance',
#                          'Dominance',
#                          'Threatening-Harassment', 
#                          'Derailing-Attacks',
#                          'Derogation-Discredit',
#                          'Derogation-Discredit',
#                          'Threatening-Harassment',
#                          'Objectification-Stereotyping',
#                          'Objectification-Stereotyping',
#                          'Derailing-Attacks',
#                          'Non']

class_map['new_class'] = ['Non',
                          'Derogation',
                          'Derogation', 
                          'Derogation',
                          'Derogation',
                          'Treatment',
                          'Treatment',
                          'Treatment',
                          'Treatment',
                          'Derogation',
                          'Derogation']

twitter_data_class = twitter_data.merge(class_map, on=['misogynous', 'misogyny_category', 'target'], how='left')

class_map
```

```python
twitter_data_class['body'] = [" ".join(filter(lambda x:x[0:2]!='RT', a.split())) for a in twitter_data_class['text']]
twitter_data_class['body'] = [" ".join(filter(lambda x:x[0]!='@', a.split())) for a in twitter_data_class['body']]
twitter_data_class['body'] = [" ".join(filter(lambda x:x[0:4]!='http', a.split())) for a in twitter_data_class['body']]
twitter_data_clean = twitter_data_class[['body', 'new_class', 'split']]
twitter_data_clean['Misogyny'] = [1 if i != 'Non' else 0 for i in twitter_data_clean.new_class]
```

```python
twitter_data_clean.to_pickle("twitter_labels.pkl")
```
